```python
import pandas as pd

movies = pd.read_csv('data/movies_w.csv')
movies.columns = ['movie_id','title','genres','year']

ratings = pd.read_csv('data/ratings_w.csv')

users = pd.read_csv('data/users_w.csv')
users.columns = ['user_id','gender','age','occupation','zip_code']
```

```python
df = pd.merge(ratings, users, how='left', on='user_id')    # 'user_id','gender','age','occupation','zip_code'
df = pd.merge(df, movies, how='left', on='movie_id')       #  'movie_id','title','genres','year'
df
```

```python
occupation = df['occupation']
occ = []

for x in occupation:
    if x==7:
        occ.append('CEO')
    elif x == 1:
        occ.append('교육자')
    elif (x == 0) | (x == 18):
        occ.append('기타')
    elif x == 8:
        occ.append('농부')
    elif x == 10:
        occ.append('미성년자')
    elif (x == 4) | (x==9) | (x==13) | (x==19):
        occ.append('비급여')
    elif (x == 2) | (x == 20) :
        occ.append('예술')
    elif (x == 12) |(x == 15)|(x == 17):
        occ.append('이공계_직장인')
    elif (x == 3) |(x == 5)|(x == 14):
        occ.append('인문계_직장인')
    elif (x == 16):
        occ.append('자영업자')
    else:
        occ.append('전문직')
    
df['occupation2'] = occ
```

```python
year = df['year']
new = []

for x in year:
    if (x >= 1919) & (x < 1920):
        new.append('1910s')
    elif (x >= 1920) & (x < 1930):
        new.append('1920s')
    elif (x >= 1930) & (x < 1940):
        new.append('1930s')
    elif (x >= 1940) & (x < 1950):
        new.append('1940s')
    elif (x >= 1950) & (x < 1960):
        new.append('1950s')
    elif (x >= 1960) & (x < 1970):
        new.append('1960s')
    elif (x >= 1970) & (x < 1980):
        new.append('1970s')
    elif (x >= 1980) & (x < 1990):
        new.append('1980s')
    elif (x >= 1990) & (x < 2000):
        new.append('1990s')
    else:
        new.append('2000s')
df['year2'] = new
```

```python
df
```

```python
gens = ['Action','Adventure','Animation',
        "Children's",'Comedy','Crime','Documentary','Drama',
        'Fantasy',   'Film-Noir',   'Horror',   'Musical',   'Mystery',   
        'Romance',   'Sci-Fi',   'Thriller',   'War'   ,'Western']
ages = ['1910s','1920s','1930s','1940s','1950s','1960s','1970s','1980s','1990s','2000s']
occ = ['CEO', '교육','기타','농부','미성년자','비급여','예술','이공계_직장인','인문계_직장인','자영업자','전문직']
```

```python
values = []

for i in gens:
    for j in ages:
        for a in occ:
            x = df[(df['genres'].str.contains(i)) & (df['year2']==j) & (df['occupation2']==a)]['rating'].mean()
            values.append((i,j,a,x))
```

```python
values
```

```python
result = pd.DataFrame(values)
result.columns = ['genres','ages','occupation','rating_mean']    # 각각 i,j,a,x 
result
```

```python
result_pivot = pd.pivot_table(result,                # 피벗할 데이터프레임
                     index = ['ages','occupation'],    # 행 위치에 들어갈 열
                     columns = 'genres',    # 열 위치에 들어갈 열
                     values = 'rating_mean')
result_pivot
```

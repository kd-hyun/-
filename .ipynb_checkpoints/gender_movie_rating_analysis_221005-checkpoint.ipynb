```python
import pandas as pd
import os
os.listdir()

movies = pd.read_csv('data/movies_w.csv')
movies.columns = ['movie_id','title','genres','year']

ratings = pd.read_csv('data/ratings_w.csv')

users = pd.read_csv('data/users_w.csv')
users.columns = ['user_id','gender','age','occupation','zip_code']
```

```python
os.getcwd()
```

```python
# 참고
[x.split('|') for x in movies['genres']]
```

```python
df = pd.merge(ratings, users, how='left', on='user_id')    # 'user_id','gender','age','occupation','zip_code'
df = pd.merge(df, movies, how='left', on='movie_id')       # 'movie_id','title','genres','year'
df
```

```python
print(df['year'].max())
print(df['year'].min())
```

```python
year = df['year']
new = []

for x in year:
    if (x >= 1919) & (x < 1920):
        new.append('1910s')
    elif (x >= 1920) & (x < 1930):
        new.append('1920s')
    elif (x >= 1930) & (x < 1940):
        new.append('1930s')
    elif (x >= 1940) & (x < 1950):
        new.append('1940s')
    elif (x >= 1950) & (x < 1960):
        new.append('1950s')
    elif (x >= 1960) & (x < 1970):
        new.append('1960s')
    elif (x >= 1970) & (x < 1980):
        new.append('1970s')
    elif (x >= 1980) & (x < 1990):
        new.append('1980s')
    elif (x >= 1990) & (x < 2000):
        new.append('1990s')
    else:
        new.append('2000s')
df['year2'] = new
```

```python
df
```

```python
gens = ['Action','Adventure','Animation',
        "Children's",'Comedy','Crime','Documentary','Drama',
        'Fantasy',   'Film-Noir',   'Horror',   'Musical',   'Mystery',   
        'Romance',   'Sci-Fi',   'Thriller',   'War'   ,'Western']
ages = ['1910s','1920s','1930s','1940s','1950s','1960s','1970s','1980s','1990s','2000s']
gender = ['M','F']
```

```python
values = []
for i in gens:
    for j in ages:
        for a in gender:
            x = df[(df['genres'].str.contains(i))&(df['year2']==j)&(df['gender']==a)]['rating'].mean()
            values.append((i,j,a,x))
            #print('genres:',i,'ages:', j,'gender:',a,'mean', x)
```

```python
values
```

```python
result = pd.DataFrame(values)
result.columns = ['genres','ages','gender','rating_mean']    # 각각 i,j,a,x 
print(result)
```

```python
result_pivot = pd.pivot_table(result,                # 피벗할 데이터프레임
                     index = ['ages','genres'],    # 행 위치에 들어갈 열
                     columns = 'gender',    # 열 위치에 들어갈 열
                     values = 'rating_mean')
result_pivot
```

```python
result_pivot.to_excel("result_gender_2.xlsx")
```

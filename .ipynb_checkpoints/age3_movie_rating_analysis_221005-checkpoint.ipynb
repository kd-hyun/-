```python
import pandas as pd
import os
os.listdir()

movies = pd.read_csv('data/movies_w.csv')
movies.columns = ['movie_id','title','genres','year']

ratings = pd.read_csv('data/ratings_w.csv')

users = pd.read_csv('data/users_w.csv')
users.columns = ['user_id','gender','age','occupation','zip_code']
```

```python
df = pd.merge(ratings, users, how='left', on='user_id')    # 'user_id','gender','age','occupation','zip_code'
df = pd.merge(df, movies, how='left', on='movie_id')       # 'movie_id','title','genres','year'
df
```

```python
year = df['year']
new = []

for x in year:
    if (x >= 1919) & (x < 1920):
        new.append('1910s')
    elif (x >= 1920) & (x < 1930):
        new.append('1920s')
    elif (x >= 1930) & (x < 1940):
        new.append('1930s')
    elif (x >= 1940) & (x < 1950):
        new.append('1940s')
    elif (x >= 1950) & (x < 1960):
        new.append('1950s')
    elif (x >= 1960) & (x < 1970):
        new.append('1960s')
    elif (x >= 1970) & (x < 1980):
        new.append('1970s')
    elif (x >= 1980) & (x < 1990):
        new.append('1980s')
    elif (x >= 1990) & (x < 2000):
        new.append('1990s')
    else:
        new.append('2000s')
df['year2'] = new
```

```python
1, 18.  미성년_사회초년생
25,35  주요_경제인구
45, 50  시니어_경제인구
56.       은퇴 
```

```python
age = df['age']
new2 = []

for i in age:
    if (i==1) | (i==18):
        new2.append('미성년_사회초년생')
    elif (i==25) | (i==35):
        new2.append('주요_경제인구')
    elif (i==45) | (i==50):
        new2.append('시니어_경제인구')
    else:
        new2.append('은퇴')
        
df['age2'] = new2
```

```python
df
```

```python
gens = ['Action','Adventure','Animation',
        "Children's",'Comedy','Crime','Documentary','Drama',
        'Fantasy',   'Film-Noir',   'Horror',   'Musical',   'Mystery',   
        'Romance',   'Sci-Fi',   'Thriller',   'War'   ,'Western']
ages = ['1910s','1920s','1930s','1940s','1950s','1960s','1970s','1980s','1990s','2000s']
age2 = ['미성년_사회초년생', '주요_경제인구','시니어_경제인구','은퇴']
```

```python
values = []
for i in gens:
    for j in ages:
        for a in age2:
            x = df[(df['genres'].str.contains(i))&(df['year2']==j)&(df['age2']==a)]['rating'].mean()
            values.append((i, j, a, x))
#             print(i, j, a, x)
```

```python
저 조건에 해당하는 dataframe을 추출한 후 rating컬럼의 평균값을 보겠다- 라는거야
```

```python
result = pd.DataFrame(values)
result.columns = ['genres','ages','age2','rating_mean']    # 각각 i,j,a,x 
print(result)
```

```python
result_pivot = pd.pivot_table(result,                # 피벗할 데이터프레임
                     index = ['ages','age2'],    # 행 위치에 들어갈 열
                     columns = 'genres',    # 열 위치에 들어갈 열
                     values = 'rating_mean')
result_pivot
```
